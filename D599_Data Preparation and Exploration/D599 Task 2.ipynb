### Import Libraries and Load Data

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset 
df = pd.read_csv('Health Insurance Dataset.csv', encoding='latin1')

# Display first few rows
df.head()
```

## A. Univariate Stats for Continuous Variables

We will be able to see

   •	 Mean, Std Dev, Min, Max
   
   •	25th, 50th (median), and 75th percentiles

```python
# Summary statistics for continuous variables
df[['age', 'bmi']].describe()
```

### Visualize Continuous Variables

```python
# Age distribution
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df['age'], kde=True, bins=20, color='skyblue')
plt.title('Age Distribution')

# BMI distribution
plt.subplot(1, 2, 2)
sns.histplot(df['bmi'], kde=True, bins=20, color='salmon')
plt.title('BMI Distribution')
plt.tight_layout()
plt.show()
```

### Univarite Stats for Categorical Variables

```python
# Frequency counts
sex_counts = df['sex'].value_counts()
smoker_counts = df['smoker'].value_counts()

print("Sex Distribution:\n", sex_counts)
print("\nSmoker Distribution:\n", smoker_counts)
```

### Visualize Categorical Variables

```python
# Bar plots for categorical variables
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
sns.countplot(x='sex', data=df, palette='pastel')
plt.title('Sex Distribution')

plt.subplot(1, 2, 2)
sns.countplot(x='smoker', data=df, palette='Set2')
plt.title('Smoker Status Distribution')

plt.tight_layout()
plt.show()
```

### Bivariate Statistics for Categorical vs. Continuous Variables

To explore how BMI varies by smoker status and sex, I used box plots. Box plots display the median, quartiles, and outliers of BMI values across each category.

```python
# Boxplot: BMI vs. Smoker status
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.boxplot(x='smoker', y='bmi', data=df, palette='Set3')
plt.title('BMI Distribution by Smoker Status')

# Boxplot: BMI vs. Sex
plt.subplot(1, 2, 2)
sns.boxplot(x='sex', y='bmi', data=df, palette='pastel')
plt.title('BMI Distribution by Sex')

plt.tight_layout()
plt.show()
```

## D. Analyzing the Data

Write python code to run the parametric test

```python
from scipy import stats

# Split the data into smokers and non-smokers
smokers = df[df['smoker'] == 'yes']
non_smokers = df[df['smoker'] == 'no']

# Conduct t-test for 'age'
age_ttest = stats.ttest_ind(smokers['age'], non_smokers['age'])

# Conduct t-test for 'bmi'
bmi_ttest = stats.ttest_ind(smokers['bmi'], non_smokers['bmi'])

# Output the results
print("T-test for Age:")
print("T-statistic:", age_ttest.statistic)
print("P-value:", age_ttest.pvalue)

print("\nT-test for BMI:")
print("T-statistic:", bmi_ttest.statistic)
print("P-value:", bmi_ttest.pvalue)
```

## Part III Non-parametric Statistical Testing

Python code using the Mann-Whitney U test to test the hypothesis

```python
# Check for NaN values in 'charges' column and remove rows with NaN values
df = df.dropna(subset=['charges'])

# Ensure the 'charges' column is numeric (in case there are non-numeric values)
df['charges'] = pd.to_numeric(df['charges'], errors='coerce')

# Now split the data into smokers and non-smokers
smokers = df[df['smoker'] == 'yes']
non_smokers = df[df['smoker'] == 'no']

# Perform the Mann-Whitney U test on health insurance costs 
u_statistic, p_value = stats.mannwhitneyu(smokers['charges'], non_smokers['charges'])

# Output the results
print("Mann-Whitney U test results:")
print("U-statistic:", u_statistic)
print("P-value:", p_value)
```
