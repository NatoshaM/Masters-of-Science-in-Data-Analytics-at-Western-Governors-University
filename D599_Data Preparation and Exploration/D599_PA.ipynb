# Check for Duplicate Entries

```python
df.duplicated()
```

```python
# Check for duplicate rows and display them if any exist

# Count duplicate rows
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# Display duplicate rows if any
if duplicate_count > 0:
    display(df[df.duplicated()])
```

```python
#create a copy of the original data frame before making changes to clean the df that way the original df remains unchanged
df_cleaned =df.copy()
```

```python
#drop duplicates in-place
df_cleaned.drop_duplicates(inplace=True)

#check that all duplicates have been sucessfully removed
print(f"Duplicate rows after removal: {df_cleaned.duplicated().sum()}")
```

# Check for Missing Values

```python
# Count missing values per column
missing_values = df_cleaned.isnull().sum()
print("Missing values per column:\n", missing_values)
```

To handle missing values for numerical columns (NumCompaniesPreviouslyWorked, AnnualProfessionalDevHrs) I will fill with meadian
and for  Categorical Column (TextMessageOptIn) I will fill with "No" if missing means the employee didn't opt in.

```python
# Fill missing values for numerical columns with median
df_cleaned['NumCompaniesPreviouslyWorked'] = df_cleaned['NumCompaniesPreviouslyWorked'].fillna(df_cleaned['NumCompaniesPreviouslyWorked'].median())
df_cleaned['AnnualProfessionalDevHrs'] = df_cleaned['AnnualProfessionalDevHrs'].fillna(df_cleaned['AnnualProfessionalDevHrs'].median())

# Fill missing values for categorical column with "No"
df_cleaned['TextMessageOptIn'] = df_cleaned['TextMessageOptIn'].fillna("No")
```

```python
#check if the output shows 0 missing values
print("Missing values per column after cleaning:\n", df_cleaned.isnull().sum())
```

# Check for Inconsistent Entries:

Inconsistent entries occur when categorical variables have typos, extra spaces, or different formats. You can check unique values in categorical columns. This helps identify spelling mistakes, extra spaces, or inconsistent capitalization (e.g., "Male" vs. "male").


```python
# Checking unique values in categorical columns for inconsistencies
categorical_columns = ['Turnover', 'CompensationType', 'JobRoleArea', 'Gender', 'MaritalStatus', 'PaycheckMethod', 'TextMessageOptIn']

for col in categorical_columns:
    print(f"Unique values in {col}:\n{df_cleaned[col].unique()}\n")
```

To fix these inconsistencies, we can use the .replace() method to standardize the values in the JobRoleArea and PaycheckMethod columns.

How it works:

.replace({old_value: new_value}) maps inconsistent values to a standard version.

Ensures all entries use a consistent format (e.g., "Human Resources" instead of "Human_Resources" or "HumanResources").

The final print() statements confirm that only the correct, standardized values remain.


```python
# Standardizing JobRoleArea values
df_cleaned['JobRoleArea'] = df_cleaned['JobRoleArea'].replace({
    'Human_Resources': 'Human Resources',
    'HumanResources': 'Human Resources',
    'InformationTechnology': 'Information Technology',
    'Information_Technology' : 'Information Technology'
})

# Standardizing PaycheckMethod values
df_cleaned['PaycheckMethod'] = df_cleaned['PaycheckMethod'].replace({
    'Mail_Check': 'Mailed Check',
    'Mail Check': 'Mailed Check',
    'MailedCheck': 'Mailed Check',
    'Direct_Deposit': 'Direct Deposit',
    'DirectDeposit': 'Direct Deposit'
})

# Verify changes
print(df_cleaned['JobRoleArea'].unique())
print(df_cleaned['PaycheckMethod'].unique())
```

# Check for Outliers



```python
(df_cleaned.columns.tolist())
```

```python
# List of numerical columns
numeric_columns = ['Age', 'Tenure', 'HourlyRate', 'HoursWeekly', 'AnnualSalary', 
                   'DrivingCommuterDistance', 'NumCompaniesPreviouslyWorked', 'AnnualProfessionalDevHrs']

# Find outliers using IQR
for col in numeric_columns:
    if col in df_cleaned.columns:  # Check if the column exists
        Q1 = df_cleaned[col].quantile(0.25)
        Q3 = df_cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = df_cleaned[(df_cleaned[col] < lower_bound) | (df_cleaned[col] > upper_bound)]
        
        print(f"Column: {col} - Outliers detected: {len(outliers)}")
    else:
        print(f"Warning: Column '{col}' not found in dataset.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Verify which numeric columns exist in df_cleaned
existing_numeric_columns = [col for col in numeric_columns if col in df_cleaned.columns]

if existing_numeric_columns:  # Proceed only if there are valid columns
    # Plot boxplots for numerical columns
    plt.figure(figsize=(12, 6))
    df_cleaned[existing_numeric_columns].boxplot(rot=45)  # Rotate labels for readability
    plt.title("Boxplots for Outlier Detection")
    plt.show()
else:
    print("Warning: No valid numerical columns found in df_cleaned.")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check if the column exists to avoid errors
if 'DrivingCommuterDistance' in df_cleaned.columns:
    plt.figure(figsize=(6, 4))
    sns.boxplot(y=df_cleaned['DrivingCommuterDistance'])
    plt.title("Boxplot for Driving Commuter Distance")
    plt.ylabel("Driving Commuter Distance")
    plt.show()
else:
    print("Error: Column 'DrivingCommuterDistance' not found in df_cleaned.")
```
